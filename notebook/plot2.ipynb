```python
%matplotlib qt
import numpy as np
import yaml
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python




for i in range(5):
    exp = str(i + 1)
    file_name = "../data/experiments/trajectory06/plan_" + exp + ".yaml"
    with open(file_name) as f:
        loadeddict = yaml.load(f)
    planned_trajectory = loadeddict.get('planned_trajectory')
    planned_trajectory = np.array(planned_trajectory)    

    file_name = "../data/experiments/trajectory06/goal_" + exp + ".yaml"
    with open(file_name) as f:
        loadeddict = yaml.load(f)
    executed_trajectory = loadeddict.get('executed_trajectory')
    executed_trajectory = np.array(executed_trajectory)

    fig = plt.figure()
    ax = Axes3D(fig)
    plt.axis('equal')
    ax.axis('equal')
    ax.set_title('trajectory_06'+exp)
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_zlabel('z-axis')
    e_i = ax.scatter(executed_trajectory[0, 0], executed_trajectory[0, 1], executed_trajectory[0, 2], label="initial_pose_executed")
    e_g = ax.scatter(executed_trajectory[-1, 0], executed_trajectory[-1, 1], executed_trajectory[-1, 2], label="goal_pose_executed") 
    p_i = ax.scatter(planned_trajectory[0, 0], planned_trajectory[0, 1], planned_trajectory[0, 2], label="initial_pose_planned")
    p_g = ax.scatter(planned_trajectory[-1, 0], planned_trajectory[-1, 1], planned_trajectory[-1, 2], label="goal_pose_planned")
    
    exe_plot, = ax.plot(executed_trajectory[:, 0], executed_trajectory[:, 1], executed_trajectory[:, 2], label='executed_trajectory')
    pla_plot, = ax.plot(planned_trajectory[:, 0], planned_trajectory[:, 1], planned_trajectory[:, 2], label='planned_trajectory')
    ax.legend([exe_plot, pla_plot, e_i, e_g, p_i, p_g], ['executed_trajectory', 'planned_trajectory', 'initial_pose_executed', 
                                                         'goal_pose_executed', 'initial_pose_planned', 'goal_pose_executed'])
    plt.show()
```

```python
def plot(name, title):
    fig = plt.figure()
    ax = Axes3D(fig)

    exe_t = []
    pla_t = []
    for i in range(5):
        exp = str(i + 1)
        file_name = name + "plan_" + exp + "_trial_1" + ".yaml"
        with open(file_name) as f:
            loadeddict = yaml.load(f)
        planned_trajectory = loadeddict.get('planned_trajectory')
        planned_trajectory = np.array(planned_trajectory)    

        file_name = name + "goal_" + exp + "_trial_1" + ".yaml"
        with open(file_name) as f:
            loadeddict = yaml.load(f)
        executed_trajectory = loadeddict.get('executed_trajectory')
        executed_trajectory = np.array(executed_trajectory)
        exe_t.append(executed_trajectory)
        pla_t.append(planned_trajectory)

    plt.axis('equal')
    ax.axis('equal')
    ax.set_title(title)
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_zlabel('z-axis')


    #1

    executed_trajectory = exe_t[0]
    planned_trajectory = pla_t[0]
    e_g_1 = ax.scatter(executed_trajectory[-1, 0], executed_trajectory[-1, 1], executed_trajectory[-1, 2], label="goal_1_goal_pose_executed") 
    p_i_1 = ax.scatter(planned_trajectory[0, 0], planned_trajectory[0, 1], planned_trajectory[0, 2], label="goal_1_initial_pose_planned")
    p_g_1 = ax.scatter(planned_trajectory[-1, 0], planned_trajectory[-1, 1], planned_trajectory[-1, 2], label="goal_1_goal_pose_planned")

    exe_plot_1, = ax.plot(executed_trajectory[:, 0], executed_trajectory[:, 1], executed_trajectory[:, 2], label='goal_1_executed_trajectory')
    pla_plot_1, = ax.plot(planned_trajectory[:, 0], planned_trajectory[:, 1], planned_trajectory[:, 2], label='goal_1_planned_trajectory')

    #2

    executed_trajectory = exe_t[1]
    planned_trajectory = pla_t[1]
    e_g_2 = ax.scatter(executed_trajectory[-1, 0], executed_trajectory[-1, 1], executed_trajectory[-1, 2], label="goal_2_goal_pose_executed") 
    p_g_2 = ax.scatter(planned_trajectory[-1, 0], planned_trajectory[-1, 1], planned_trajectory[-1, 2], label="goal_2_goal_pose_planned")

    exe_plot_2, = ax.plot(executed_trajectory[:, 0], executed_trajectory[:, 1], executed_trajectory[:, 2], label='goal_2_executed_trajectory')
    pla_plot_2, = ax.plot(planned_trajectory[:, 0], planned_trajectory[:, 1], planned_trajectory[:, 2], label='goal_2_planned_trajectory')

    #3
    executed_trajectory = exe_t[2]
    planned_trajectory = pla_t[2]
    e_g_3 = ax.scatter(executed_trajectory[-1, 0], executed_trajectory[-1, 1], executed_trajectory[-1, 2], label="goal_3_goal_pose_executed") 
    p_g_3 = ax.scatter(planned_trajectory[-1, 0], planned_trajectory[-1, 1], planned_trajectory[-1, 2], label="goal_3_goal_pose_planned")

    exe_plot_3, = ax.plot(executed_trajectory[:, 0], executed_trajectory[:, 1], executed_trajectory[:, 2], label='goal_3_executed_trajectory')
    pla_plot_3, = ax.plot(planned_trajectory[:, 0], planned_trajectory[:, 1], planned_trajectory[:, 2], label='goal_3_planned_trajectory')

    #4
    executed_trajectory = exe_t[3]
    planned_trajectory = pla_t[3]
    e_g_4 = ax.scatter(executed_trajectory[-1, 0], executed_trajectory[-1, 1], executed_trajectory[-1, 2], label="goal_4_goal_pose_executed") 
    p_g_4 = ax.scatter(planned_trajectory[-1, 0], planned_trajectory[-1, 1], planned_trajectory[-1, 2], label="goal_4_goal_pose_planned")

    exe_plot_4, = ax.plot(executed_trajectory[:, 0], executed_trajectory[:, 1], executed_trajectory[:, 2], label='goal_4_executed_trajectory')
    pla_plot_4, = ax.plot(planned_trajectory[:, 0], planned_trajectory[:, 1], planned_trajectory[:, 2], label='goal_4_planned_trajectory')
    
    #5
    executed_trajectory = exe_t[4]
    planned_trajectory = pla_t[4]
    e_g_5 = ax.scatter(executed_trajectory[-1, 0], executed_trajectory[-1, 1], executed_trajectory[-1, 2], label="goal_5_goal_pose_executed") 
    p_g_5 = ax.scatter(planned_trajectory[-1, 0], planned_trajectory[-1, 1], planned_trajectory[-1, 2], label="goal_5_goal_pose_planned")

    exe_plot_5, = ax.plot(executed_trajectory[:, 0], executed_trajectory[:, 1], executed_trajectory[:, 2], label='goal_5_executed_trajectory')
    pla_plot_5, = ax.plot(planned_trajectory[:, 0], planned_trajectory[:, 1], planned_trajectory[:, 2], label='goal_5_planned_trajectory')
    
    ax.legend([exe_plot_1, pla_plot_1, exe_plot_2, pla_plot_2, exe_plot_3, pla_plot_3, exe_plot_4, pla_plot_4, \
               exe_plot_5, pla_plot_5, p_i_1, p_g_1, e_g_2, p_g_2, e_g_3, p_g_3, e_g_4, p_g_4, e_g_5, p_g_5], \
             ['goal_1_executed_trajectory', 'goal_1_planned_trajectory', 'goal_2_executed_trajectory', 'goal_2_planned_trajectory', \
             'goal_3_executed_trajectory', 'goal_3_planned_trajectory', 'goal_4_executed_trajectory', 'goal_4_planned_trajectory',\
             'goal_5_executed_trajectory', 'goal_5_planned_trajectory', 'start pose'])
    plt.show()
```
