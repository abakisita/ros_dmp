```python
#############################################################

from os import listdir
from os.path import isfile, join
mypath = "/home/abhishek/abhishek_logs"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]


files = []
for fe in onlyfiles:
    if fe[0] == "e":
        fp = "planned" + fe[8:]
        files.append([fe, fp])
        

a = [21, 13, 26, 27, 38, 40]        

j = 1
error_list = []

for i in a:
    
    fi = files[i]
    fig = plt.figure()
    ax = Axes3D(fig)
    file_name = "/home/abhishek/abhishek_logs/" + fi[1]
    with open(file_name) as f:
        loadeddict = yaml.load(f)
    planned_trajectory = loadeddict.get('planned_trajectory')
    planned_trajectory = np.array(planned_trajectory)    

    file_name = "/home/abhishek/abhishek_logs/" + fi[0]
    with open(file_name) as f:
        loadeddict = yaml.load(f)
    executed_trajectory = loadeddict.get('executed_trajectory')
    executed_trajectory = np.array(executed_trajectory)

    ax.set_title("Object grasp " + str(j))
    ax.plot(planned_trajectory[:,0], planned_trajectory[:,1], planned_trajectory[:,2])
    ax.plot(executed_trajectory[:,0], executed_trajectory[:,1], executed_trajectory[:,2])
    
    ax.scatter(planned_trajectory[0,0], planned_trajectory[0,1], planned_trajectory[0,2])
    ax.scatter(planned_trajectory[-1,0], planned_trajectory[-1,1], planned_trajectory[-1,2])
    ax.scatter(executed_trajectory[-1,0], executed_trajectory[-1,1], executed_trajectory[-1,2])
    
    ax.legend(["planned_trajectory", "executed_trajectory", "initial position", "goal_position_planned", "goal_position_executed"])
    j += 1
    ax.set_xlabel('x (meter)')
    ax.set_ylabel('y (meter)')
    ax.set_zlabel('z (meter)')
    error = 0.0
    for k in range(executed_trajectory.shape[1]):
        dist = []
        for l in range(planned_trajectory.shape[1]):
            
            dist.append(np.linalg.norm((executed_trajectory[k,:] - planned_trajectory[l,:])))
        
        error += min(dist)
    error_list.append(error)
plt.show()
print(error_list)
```

```python

plt.figure()

plt.scatter(np.arange(1,7), error_list, c='black')
plt.xlabel('Number of trials')
plt.ylabel('Normalized point to point error (meter)')
plt.title("Error in trajectory execution")
plt.plot()
        
```
